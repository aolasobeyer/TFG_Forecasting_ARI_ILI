## Loading ARI and ILI incidence data

The datasets are stored locally in CSV format and include the following key variables:

- `location`: ISO-2 country code  
- `truth_date`: end date of the epidemiological week (Sunday)  
- `year_week`: epidemiological year and week  
- `value`: incidence per population unit

The dates are converted to proper datetime objects to facilitate time series analysis in later stages.


```python
import pandas as pd
from pathlib import Path

# notebook directory
BASE_DIR = Path.cwd()

# construct paths
DATA_DIR = BASE_DIR.parent / "data"

ari_df = pd.read_csv(DATA_DIR / "latest-ARI_incidence.csv")
ili_df = pd.read_csv(DATA_DIR / "latest-ILI_incidence.csv")

ari_df.head()
```
